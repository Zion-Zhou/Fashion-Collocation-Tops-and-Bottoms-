```python
from datetime import datetime
from itertools import dropwhile, takewhile

import instaloader

L = instaloader.Instaloader(download_videos=False, download_video_thumbnails=False, 
                            download_geotags=False, download_comments=False, 
                            save_metadata=False, compress_json=False,
                           post_metadata_txt_pattern='')
```

# based on User


```python
username = input("username: ")
posts = instaloader.Profile.from_username(L.context, username).get_posts()

SINCE = datetime(2020, 4, 11)
UNTIL = datetime(2018, 1, 1)

cnt = 0
for post in takewhile(lambda p: p.date > UNTIL, dropwhile(lambda p: p.date > SINCE, posts)):
    # print(post.date)
    cnt+=1
    L.download_post(post, username)
print("{}: {} posts".format(username, cnt))
```

```python
posts = L.get_hashtag_posts('ootdguide') 
cnt = 0
SINCE = datetime(2020, 4, 11)
UNTIL = datetime(2018, 1, 1)

for post in takewhile(lambda p: p.date > UNTIL, dropwhile(lambda p: p.date > SINCE, posts)):
    cnt+=1
    print(post.date)
    L.download_post(post, '#ootdguide')
print("{} post".format(cnt))
```

```
from PIL import Image
import os,shutil
from tqdm import tqdm

def compressImage(srcPath,dstPath):
    for filename in tqdm(os.listdir(srcPath)):
        if not os.path.exists(dstPath):
                os.makedirs(dstPath)
 
        srcFile=os.path.join(srcPath,filename)
        dstFile=os.path.join(dstPath,filename)
 
        if os.path.isfile(srcFile):
            try:
                
                sImg=Image.open(srcFile)
                w,h=sImg.size
                dImg=sImg.resize((int(400),int(600)),Image.ANTIALIAS) # size is customized
                dImg.save(dstFile) 
            except Exception:
                print(dstFile+"failed")
 
        if os.path.isdir(srcFile):
            compressImage(srcFile, dstFile)
```

```
compressImage("./prepare","./compress") # prepare folder is where you put the image and the code will put all the compressed image in the compress folder
```
